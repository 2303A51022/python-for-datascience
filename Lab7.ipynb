```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Load Dataset
# -----------------------------
# Example sales_data.csv content:
# Quarter,Sales
# Q1,15000
# Q2,20000
# Q3,18000
# Q4,25000

df = pd.read_csv("sales_data.csv")

# -----------------------------
# Step 2: Plot Line Chart
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Sales"], marker='o', linestyle='-', color='blue', label="Sales")

# Highlight peak sales
peak_idx = df["Sales"].idxmax()
plt.scatter(df["Quarter"][peak_idx], df["Sales"][peak_idx], color="red", s=100, label="Peak Sales")
plt.text(df["Quarter"][peak_idx], df["Sales"][peak_idx]+500, "Peak", ha="center", fontsize=10, color="red")

# -----------------------------
# Step 3: Formatting the Chart
# -----------------------------
plt.title("Quarterly Sales Performance", fontsize=14, fontweight="bold")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

# Show chart
plt.show()
```

```python
# Create a dummy sales_data.csv file
with open("sales_data.csv", "w") as f:
    f.write("Quarter,Sales\n")
    f.write("Q1,15000\n")
    f.write("Q2,20000\n")
    f.write("Q3,18000\n")
    f.write("Q4,25000\n")
```
